```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

import random
import sys 

import tensorly as tl
from tensorly.decomposition import non_negative_parafac
import scipy.sparse
```

```python
import pickle as pkl

samples = 10 
rank = 2 
comm = 5 
# ind = 3 #int(sys.argv[3])
factor_list = []
path = '/media/garner1/hdd1/gpseq/10000'
for ind in range(1,11):
    fileName = path + '/NNparafac' + '_rank' + str(rank) + 'sample' + str(ind) + 'of' + str(samples) + '.pkl'
    fileObject = open(fileName, 'rb')
    factor_list.append(pkl.load(fileObject))
    fileObject.close()
```

```python
fidelity = np.zeros((len(factor_list),len(factor_list),rank,rank))
for f1 in range(len(factor_list)):
    for f2 in range(f1,len(factor_list)):
        for comm1 in range(rank):
            A1 = factor_list[f1][1][:,comm1]/tl.norm(factor_list[f1][1][:,comm1])
            B1 = factor_list[f1][2][:,comm1]/tl.norm(factor_list[f1][2][:,comm1])
            C1 = factor_list[f1][0][:,comm1]/tl.norm(factor_list[f1][0][:,comm1])
            for comm2 in range(comm1,rank):
                A2 = factor_list[f2][1][:,comm2]/tl.norm(factor_list[f2][1][:,comm2])
                B2 = factor_list[f2][2][:,comm2]/tl.norm(factor_list[f2][2][:,comm2])
                C2 = factor_list[f2][0][:,comm2]/tl.norm(factor_list[f2][0][:,comm2])
                fidelity[f1,f2,comm1,comm2] = np.dot(A1,A2)*np.dot(B1,B2)*np.dot(C1,C2)
                fidelity[f1,f2,comm2,comm1] = fidelity[f1,f2,comm1,comm2]
                fidelity[f2,f1,comm2,comm1] = fidelity[f1,f2,comm1,comm2]
                fidelity[f2,f1,comm1,comm2] = fidelity[f1,f2,comm1,comm2]
                if comm1 == comm2 and f1 == f2: fidelity[f1,f2,comm1,comm2] = 0
#         else: print(comm1,comm2,fidelity[comm2,comm1])
#         if fidelity[comm1,comm2]>1e-01: print(comm1,comm2,fidelity[comm2,comm1])
```

```python
'''
Show the distribution of fidelities between all community pairs
'''
plt.hist(fidelity[fidelity>0.1].flatten(),bins='auto')
```

```python
f1 = 1
f2 = 9
comm1_list = range(rank)
comm2_list = np.argmax(fidelity[f1,f2,:,:],axis=1)
for i in range(rank)[:2]:
    fig, ax = plt.subplots(1, 2, sharex='col', sharey='row',figsize=(30,30))
    a = factor_list[f1][1][:,comm1_list[i]]
    b = factor_list[f1][2][:,comm1_list[i]]
    mat_1 = np.outer(a,b)
    a = factor_list[f2][1][:,comm2_list[i]]
    b = factor_list[f2][2][:,comm2_list[i]]
    mat_2 = np.outer(a,b)
    print(str(fidelity[f1,f2,comm1_list[i],comm2_list[i]]))
    ax[0].imshow(mat_1, cmap='Blues', interpolation='nearest')
    ax[1].imshow(mat_2, cmap='Blues', interpolation='nearest')
    plt.show()    
```
