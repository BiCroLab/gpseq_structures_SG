```python
import os
import csv
import numpy as np
import random
import tensorly as tl
from tensorly.decomposition import non_negative_parafac
import scipy.sparse
import sys
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

samples = 100 #numb of sampled 3d structured

path = '/media/garner1/hdd1/gpseq/10000'
# path = '/media/garner1/hdd1/gpseq/10000G'

files = os.listdir(path) # dir list in path with different configurations
config_sample = random.sample(files, k=samples) # sample k times without replacement from configurations
```

```python
rank = 5 # numb of components in svd. The greater this value the slower the parafac convergence 
comm = 5 # numb of communities to retrive
```

```python
T = np.zeros(shape = (samples, 3043, 3043), dtype = np.float32)
graph_idx = 0
for config in config_sample:
    dirname = os.fsdecode(config)
    filename = os.path.join(path, dirname+'/coords.csv_tsvd.npz')
    if os.path.isfile(filename): 
        svd = np.load(filename)
        u = svd['u'][:,:rank]
        s = svd['s'][:rank]
        vt = svd['vt'][:rank,:]
        T[graph_idx,:,:] = np.dot(np.dot(u,np.diag(s)),vt)
        del svd
        continue
    else:
        T[graph_idx,:,:] = np.zeros(shape = (3043,3043),)
        continue
    graph_idx += 1

print(T.shape)   
factors = non_negative_parafac(T, rank=comm, verbose=1, n_iter_max=20,tol=1e-08)
[factors[1][ind].shape for ind in range(3)]
```

```python
for i in range(comm):
    a = factors[1][1][:,i]
    b = factors[1][2][:,i]
    c = factors[1][0][:,i]
    mat = np.outer(0.5*(a+b),0.5*(a+b)) # symmetrize wrt a & b
    print(tl.norm(a,2)*tl.norm(b,2)*tl.norm(c,2))
#     plt.imshow(mat, cmap='Blues', interpolation='nearest')
    plt.figure(figsize=(10, 10))
    sns.heatmap(mat,cmap='Blues',square=True,xticklabels=False,yticklabels=False)
    plt.show()  
      
```

```python
'''Strenght of the communities in different SC'''
strength = np.zeros(shape=factors[1][0].shape)
for k in range(samples):
    for r in range(comm):
        strength[k,r] = factors[1][0][k,r]*np.sum(factors[1][1][:,r])
strength[0,:]=0
# strength[:,0]=0

plt.figure(figsize=(10, 30))
sns.heatmap(strength,cmap='Blues',square=True,xticklabels=False,yticklabels=False)
plt.show()  
```

```python
'''Create a normalized community membership matrix'''
S = np.zeros(shape=factors[1][2].shape)
norma = np.sum(factors[1][2],axis=0)
for c in range(comm):
    S[:,c] = factors[1][2][:,c]/norma[c]
```

```python
# path = '/media/garner1/hdd1/gpseq/10000'
path = '/media/garner1/hdd1/gpseq/10000G'
new_sample = random.sample(files, k=100) # sample k times without replacement from configurations
```

```python
# new_sample = config_sample
modularity_values = []
for config in new_sample:
    dirname = os.fsdecode(config)
    filename = os.path.join(path, dirname+'/coords.csv_sparse_graph.npz')
    if os.path.isfile(filename): 
        A = scipy.sparse.load_npz(filename)
        '''construct modularity matrix'''
        M = A
        k = A.sum(axis=0)
        w = A.sum(axis=None)
        M = A - np.outer(k,k)*0.5/w
        modularity_value = np.trace(np.dot(np.dot(S.transpose(),M),S))/(2.0*w)
        modularity_values.append(modularity_value)
        continue
    else:
        continue
```
